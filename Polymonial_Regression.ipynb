```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from matplotlib.pyplot import scatter
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
X=6*np.random.rand(200,1)-3
y=0.8*X**2+0.9*X+2+np.random.randn(200,1)
```

```python
plt.plot(X,y,'b.')
plt.show
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
lr =LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
plt.plot(X_test,y_pred,'r-')
plt.plot(X,y,'b.')
plt.show
```

```python
poly=PolynomialFeatures(degree=2,include_bias=True)
```

```python
X_train_trans=poly.fit_transform(X_train)
X_test_trans=poly.fit_transform(X_test)
```

```python
print(X_train[0])
print(X_train_trans[0])
```

```python
lr.fit(X_train_trans,y_train)
```

```python
y_pred=lr.predict(X_test_trans)
```

```python
r2_score(y_test,y_pred)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

3D POLYMONIAL REGRESSION


```python
x=7*np.random.rand(200,1)-2.8
y=7*np.random.rand(200,1)-2.8
z=x**2+y**2+0.2*x+0.2*y+0.1*x*y+2+np.random.randn(200,1)
```

```python
import plotly.express as px
df=px.data.iris()
fig=px.scatter_3d(df,x=x.ravel(),y=y.ravel(),z=z.ravel())
fig.show()
```

```python
# Combine x and y into a single array with shape (200, 2)
X_combined = np.hstack((x, y))

# Fit the linear regression model
lr.fit(X_combined, z)

x_input = np.linspace(x.min(), x.max(), 10)
y_input = np.linspace(y.min(), y.max(), 10)
xGrid, yGrid = np.meshgrid(x_input,y_input)

# Combine the meshgrid points for prediction
final = np.vstack((xGrid.ravel(), yGrid.ravel())).T

z_final = lr.predict(final).reshape(10,10)
```

```python
import plotly.graph_objects as go

fig = px.scatter_3d(df, x=x.ravel(), y=y.ravel(), z=z.ravel())

fig.add_trace(go.Surface(x = x_input, y = y_input, z =z_final ))

fig.show()
```
